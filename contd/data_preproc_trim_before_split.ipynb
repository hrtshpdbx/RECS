```python
data=pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2009_public.csv")
data
```

```python
# Select only numerical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
cat_cols = data.select_dtypes(include=['object']).columns

num_data= data[numerical_cols]
cat_data= data[cat_cols]
num_data.shape,cat_data.shape
```

```python
num_data.head()
```

```python
columns_to_check = num_data.columns.tolist()
```

```python
# Make a copy of the DataFrame columns
columns_to_check = num_data.columns.tolist()
cols_dropped=[]
# Iterate through each column in the DataFrame
i=0
while i< len(columns_to_check):
    col_x=columns_to_check[i]

    # Create a list to store columns to drop
    cols_to_drop = []

    # Iterate through each column (except col_x)
    for col_y in columns_to_check:
        if col_y != col_x:
            # Calculate correlation between col_x and col_y
            corr = num_data[col_x].corr(num_data[col_y])

            # If correlation is higher than 0.95, add col_y to cols_to_drop
            if corr > 0.95:
                cols_to_drop.append(col_y)

    # Drop highly correlated columns from the DataFrame
    if len(cols_to_drop) > 0:
        # print(f"Removing columns highly correlated with {col_x}: {cols_to_drop}")
        num_data.drop(columns=cols_to_drop, inplace=True)
        # Update columns_to_check after dropping highly correlated columns
        columns_to_check = [col for col in columns_to_check if col not in cols_to_drop]
        cols_dropped.extend(cols_to_drop)

    if (i%15 == 0):
        print(i/len(columns_to_check))
    i+=1
```

```python
print(num_data.shape, cols_dropped)
```

```python
col_target = 'KWH'

# Calculate correlation of each feature column with the target column
correlation_with_target = num_data.corr()[col_target].abs()

# Drop columns whose absolute correlation with the target column is less than 10%
columns_to_drop = correlation_with_target[correlation_with_target < 0.1].index

# Drop the columns from the DataFrame
num_data.drop(columns=columns_to_drop, inplace=True)


num_data.shape
```

```python
num_data
```

```python
layout=pd.read_csv("/content/drive/MyDrive/RECS_stuff/public_layout_2009.csv")
```

```python
mapping = dict(zip(layout['Variable Name'], layout['Variable Label']))
cat_data = cat_data.rename(columns=mapping)
num_data = num_data.rename(columns=mapping)
data= data.rename(columns=mapping)
```

```python
cols=list(num_data.columns)
data= data[cols]
data = data.loc[:, ~data.columns.duplicated()]
data.head()
```

```python
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
len(num_cols)
```

```python
nonsense= [str(col) for col in num_cols if (data[col]==-2).mean()>=0.1]     #we're checking if there's more than 10% -2's in a col, if yes, drop
#-2 is a value we see very often. there's a very high chance it means missing value because in columns age group of 5th, 6th, 7th, 8th member of the family >97% of values were -2.
len(nonsense)
```

```python
data.drop(columns=nonsense, inplace=True)
data.head()
```

We're gonna add division and typehuq and divide dataset by possible combinations

```python
data_2009_trimmed = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2009_public_trimmed.csv")
data_2015_trimmed = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2015_public_trimmed.csv")
data_2020_trimmed = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2020_public_trimmed.csv")
```

```python
data_2009 = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2009_public.csv")
data_2015 = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2015_public_v4.csv")
data_2020 = pd.read_csv("/content/drive/MyDrive/RECS_stuff/recs2020_public_v7.csv")
```

```python
print("Number of columns:")
print(data_2009_trimmed.shape[1],data_2015_trimmed.shape[1],data_2020_trimmed.shape[1])
```

```python
dataframes_trimmed = [data_2009_trimmed, data_2015_trimmed, data_2020_trimmed]
dataframes = [data_2009, data_2015, data_2020]
for dataframe_trimmed,dataframe in zip(dataframes_trimmed,dataframes):
  if 'DIVISION' not in dataframe_trimmed.columns:
    dataframe_trimmed['DIVISION'] = dataframe['DIVISION']
  if 'TYPEHUQ' not in dataframe_trimmed.columns:
    dataframe_trimmed['TYPEHUQ'] = dataframe['TYPEHUQ']
```

```python
print(data_2009_trimmed.shape[1],data_2015_trimmed.shape[1],data_2020_trimmed.shape[1])
```

In 2009, division and housing unit type aren't considered important

In 2015, division isn't considere important

In 2020, both are considered important

```python
print(data_2009_trimmed['DIVISION'].dtype,data_2015_trimmed['DIVISION'].dtype,data_2020_trimmed['DIVISION'].dtype)
print(data_2009_trimmed['TYPEHUQ'].dtype,data_2015_trimmed['TYPEHUQ'].dtype,data_2020_trimmed['TYPEHUQ'].dtype)
```

We'll map at the very end:

typehuq for all:

mobile home, single-family, detached, single-family attached, apartment in building with 2-4 units, apartment in building with 5+ units

2009 and 2015:

Divisions: new england, middle atlantic, east north central, west north central, south atlantic, east south central, west south central, mountain north, mountain south, pacific

2020:

divisions: east north central, east south central, middle atlantic, mountain north, mountain south, new england, pacific, south atlantic, west north central, west south central


```python
print(data_2009_trimmed['DIVISION'],data_2015_trimmed['DIVISION'],data_2020_trimmed['DIVISION'])
print(data_2009_trimmed['TYPEHUQ'],data_2015_trimmed['TYPEHUQ'],data_2020_trimmed['TYPEHUQ'])
```

```python
import os

# Directory in Google Drive where files will be saved
output_dir = "/content/drive/MyDrive/RECS_stuff/split_datasets"  # Update this to your desired location

# Create directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Create a dictionary mapping names to DataFrames
dataframes_trimmed = {
    "data_2009_trimmed": data_2009_trimmed,
    "data_2015_trimmed": data_2015_trimmed,
    "data_2020_trimmed": data_2020_trimmed,
}

# Iterate over all unique combinations of DIVISION and TYPEHUQ
for name, dataframe_trimmed in dataframes_trimmed.items():
    for division in dataframe_trimmed['DIVISION'].unique():
        for typehuq in dataframe_trimmed['TYPEHUQ'].unique():
            # Filter the dataset for the specific combination
            subset = dataframe_trimmed[
                (dataframe_trimmed['DIVISION'] == division) &
                (dataframe_trimmed['TYPEHUQ'] == typehuq)
            ]

            # If subset is not empty, save it
            if not subset.empty:
                filename = f"{name}_division{division}_typehuq{typehuq}.csv"
                filepath = os.path.join(output_dir, filename)
                subset.to_csv(filepath, index=False)
                print(f"Saved: {filepath}")
```
